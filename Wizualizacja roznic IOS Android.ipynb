```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import FormatStrFormatter

# Wczytaj dane
acelMich = pd.read_csv('Michal/AccelerometerUncalibrated.csv').drop(columns=['time', 'seconds_elapsed'])
acelKasper = pd.read_csv('Kasper/AccelerometerUncalibrated.csv').drop(columns=['time', 'seconds_elapsed'])

# Utwórz folder, jeśli nie istnieje
import os
if not os.path.exists('output_plots'):
    os.makedirs('output_plots')

# Utwórz wykresy i zapisz jako pliki JPEG
for i in range(1, 7):
    plt.figure(figsize=(10, 5), dpi=200)

    if i == 1:
        plt.plot(acelKasper['x'] * 9.81)
        plt.grid()
        plt.title('IOS oś x')
        plt.ylim(-4, 14.5)
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    elif i == 2:
        plt.plot(acelMich['x'])
        plt.grid()
        plt.title('android oś x')
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.ylim(-4, 14.5)
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))
    elif i == 3:
        plt.plot(acelKasper['y'] * 9.81)
        plt.grid()
        plt.title('IOS oś y')
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.ylim(-12, 12)
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))# Ustaw zakres osi y
    elif i == 4:
        plt.plot(acelMich['y'])
        plt.grid()
        plt.title('android oś y')
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.ylim(-12, 12)
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))# Ustaw zakres osi y
    elif i == 5:
        plt.plot(acelKasper['z'] * 9.81)
        plt.grid()
        plt.title('IOS oś z')
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.ylim(-2.5, 22.5)
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))# Ustaw zakres osi y
    elif i == 6:
        plt.plot(acelMich['z'])
        plt.grid()
        plt.title('android oś z')
        plt.xlabel("próbki")
        plt.ylabel(r'$m/s^2$')
        plt.ylim(-2.5, 22.5)
        plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))# Ustaw zakres osi y
```
